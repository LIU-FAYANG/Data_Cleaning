```python
import csv
import os
```

```python
try:
    if not os.path.exists('txt'):
        os.makedirs('txt')
except OSError:
    print ('Error: Creating directory of data')




with open("updated_cleaned.csv", "r", ) as f:
    reader = csv.reader(f)

    rownumber = 0
    for row in reader:
        g=open('txt'+ '/'+str(rownumber-1)+".txt","w")
        for i in row:
            if(i == 'x'):
                g.write('')
            elif(i!=''):
                g.write(i + ' ')
            else:
                g.write('\n')
        rownumber = rownumber + 1
        g.close()
os.remove("txt/-1.txt")
```
